```python
driver = webdriver.Chrome("D:\chromedriver/chromedriver.exe")
```

```python
def move_page(searching, page):
    url = f'https://search.naver.com/search.naver?query={searching}&nso=&where=blog&sm=tab_viw.all&start={page}'
    
    return url
# f'https://search.naver.com/search.naver?where=article&sm=tab_viw.blog&query={searching}&start={page}'
# f'https://search.naver.com/search.naver?where=blog&sm=tab_viw.blog&query={searching}&start={page}''
```

```python
def get_data2(driver):
    driver.switch_to.frame('mainFrame')
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    try:
        #title = driver.find_elements_by_class_name('.se-fs-.se-ff-')[0].text
        title = soup.select('span.se-fs-.se-ff-')[0].text
    except Exception as e:
        title = ''
        print('예외가 발생했습니다.1', e)
    '''
    try:
        views = soup.select('span.count')[0].text.replace('조회 ', '')
    except Exception as e:
        views = 0
        print('예외가 발생했습니다.2', e)
    '''   
    #datetime = driver.find_elements_by_class_name('.se_publishDate.pcol2')[0].text
    datetime = soup.select('span.se_publishDate.pcol2')[0].text
    date = datetime[:10]
    
    try:
        #content = driver.find_elements_by_class_name('.se-main-container')[0].text
        content = soup.select('div.se-main-container')[0].text
    except Exception as e: 
        content = ''
        print('예외가 발생했습니다.3', e)
    
    try:
        #like = re.findall("\d+", driver.find_elements_by_class_name('.u_cnt._count')[0].text)[0]
        like = re.findall("\d+", soup.select('em.u_cnt._count')[0].text)[0]
    except Exception as e:
        like = 0
        print('예외가 발생했습니다.4', e)
    '''
    bt = driver.find_elements_by_css_selector('#printPost1 > tbody > tr > td.bcc > div.post-btn.post_btn2 > div.wrap_postcomment > div.area_comment.pcol2')
    bt = bt.pop()
    try:
        bt.click()
    except Exception as e:
        print('예외가 발생했습니다....', e)
                
    time.sleep(3 + random.uniform(1, 2))
    
    all_reply = soup.select('ul.u_cbox_list')
    reply_list = []
    
    for reply in all_reply:
        reply = reply.text.strip()
        reply_list.append(reply)
    
    seperate_mark = '|||||'
    reply_joined = seperate_mark.join(reply_list)
    '''
    data = {"title":title, "date": date, "content":content, "likes":like}
    
    return data
```

```python
def ceawling(searching):
    p = 1
    for page in tqdm(range(1, 36)):
        url = move_page(searching, p)
        p += 30
        
        driver.get(url)
        
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        time.sleep(2)
        
        article = driver.find_elements_by_class_name('api_txt_lines.total_tit')
        for ar in article:
            try:
                ar.click()
            except Exception as e:
                print('예외가 발생했습니다....', e)
                
            time.sleep(3 + random.uniform(1, 2))
            
            change_tab = driver.window_handles[-1]
            driver.switch_to.window(change_tab)
        
            try:
                data = get_data2(driver)
                result.append(data)
            
            except Exception as e:
                print('예외가 발생했습니다.', e)
            
            driver.close()
        
            change_tab = driver.window_handles[-1]
            driver.switch_to.window(change_tab)
```

```python
result = []
searching = '알파트라디올'

ceawling(searching)
```
