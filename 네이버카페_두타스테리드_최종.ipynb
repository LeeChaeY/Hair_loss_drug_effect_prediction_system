```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import pandas as pd
from tqdm.notebook import tqdm
import random
from selenium.webdriver.common.keys import Keys
import numpy as np
import re
#스크롤 크롤링하기
```

```python
driver = webdriver.Chrome("D:\chromedriver/chromedriver.exe")
```

```python
def move_page(searching, page):
    url = f'https://search.naver.com/search.naver?where=article&sm=tab_viw.blog&query={searching}&start={page}'
    
    return url
# f'https://search.naver.com/search.naver?where=article&sm=tab_viw.blog&query={searching}&start={page}'
# f'https://search.naver.com/search.naver?where=blog&sm=tab_viw.blog&query={searching}&start={page}''
```

```python
def get_data(driver):
    driver.switch_to.frame('cafe_main')
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    try:
        title = soup.select('h3.title_text')[0].text
    except Exception as e:
        title = ''
        print('예외가 발생했습니다.1', e)
    
    try:
        views = soup.select('span.count')[0].text.replace('조회 ', '')
    except Exception as e:
        views = '0'
        print('예외가 발생했습니다.2', e)
        
    datetime = soup.select('span.date')[0].text
    date = datetime[:10]
    
    try:
        content = soup.select('.article_container')[0].text #se-module se-module-text
    except Exception as e: 
        content = ''
        print('예외가 발생했습니다.3', e)
    
    try:
        like = re.findall("\d+", soup.select('em.u_cnt._count')[0].text)[0]
    except Exception as e:
        like = 0
        print('예외가 발생했습니다.4', e)
    
    all_reply = soup.select('span.text_comment')
    reply_list = []
    
    for reply in all_reply:
        reply = reply.text.strip()
        reply_list.append(reply)
    
    seperate_mark = '|||||'
    reply_joined = seperate_mark.join(reply_list)
    
    data = {"title":title, "views":views, "date": date, "content":content, "likes":like, "reply":reply_joined}
    
    return data
```

```python
def ceawling(searching):
    p = 1
    for page in tqdm(range(1, 102)):
        url = move_page(searching, p)
        p += 10
    
        driver.get(url)
        time.sleep(2)
        
        for article in range(1, 11):
            ar = driver.find_elements_by_css_selector('#_view_review_body_html > div > more-contents > div > ul > li:nth-child('+ str(article)+') > div > a')
            ar = ar.pop()
            
            try:
                ar.click()
            except Exception as e:
                print('예외가 발생했습니다....', e)
                
            time.sleep(3 + random.uniform(1, 2))
            
            change_tab = driver.window_handles[-1]
            driver.switch_to.window(change_tab)
        
            try:
                data = get_data(driver)
                result.append(data)
            
            except Exception as e:
                print('예외가 발생했습니다.', e)
            
            driver.close()
        
            change_tab = driver.window_handles[-1]
            driver.switch_to.window(change_tab)
```

```python
result = []
searching = '피나스테리드'

ceawling(searching)
```
